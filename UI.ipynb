```python
def factorial_number():
        
    user_input= int (input ('Enter a number: '))
    print('Your entered number is ', user_input)
    factorial = 1
    for i in range (1, user_input+1):
        factorial = factorial *i
    print('Factorial of the', user_input,'is: ', factorial)


def countdown(minutes, seconds):
        
    import countdown
    # Get user input for minutes and seconds

    minutes = int(input("Enter the number of minutes: "))

    seconds = int(input("Enter the number of seconds: "))

    countdown(minutes, seconds)
    
def calculate_cube(number):
    cube=number **3
    return cube
    num=int(input('Enter a Number: '))
    result=calculate_cube(num)
    print(f'The cube of {num} is {result}')
    
def dice_game():
    #Welcome To The Dice Game
    print('Welcome into The Dice Game')
    #User Name
    name=input('Enter Your Name')
    # Greet The User
    print('Hello',name)
    #Input_User
    user_input=int(input('Enter The Number 1 to 6:'))
    import random
    result=random.randrange(1,7)
    print('You Have Chose',user_input)
    print('Dice Result',result)
    if result==user_input:
        print('*'*30)
        print('You Win')
        print('*'*30)
    else:
        print('*'*30)
        print('You Lose')
        print('*'*30)

def welcome_to_EPL():
    print ('Welcome to EPL')
    name=input('Enter your name here ')
    print('Hello', name)
    budget=input('Enter your ticket budget here')
    print('Your budget is', budget)

    if int(budget)>=5000:
        print('You can afford E Zone!')
    elif int(budget)>=2500 and int(budget)<=5000:
        print('You can afford C Zone!')
    else:
        print('Please watch it on TV')
        
    
def ticket_booking():
    print('Welcome to Ticket Booking Desk')
    name=input('Tell me your name ')
    print('Hello', name)
    print(' Our 4 types sessions')
    print('1:250 to 500')
    print('2:100 to 250')
    print('3:50 to 100')
    print('4:50')
    budget=input('Enter your ticket budget ')
    if int(budget)>=500:
        print('Ola Se ja')
    elif int(budget)>=250 and int(budget)<500:
        print('Auto se ja') 
    elif int(budget)>=100 and int(budget)<250:
        print('Train Se ja')
    elif int(budget)>=50 and int(budget)<100:
            print('Bus se ja')
    else:
        pass

def odd_even_number():
    num=int(input('Enter number here'))
    if num%2==0:
        print('Numver is Even')
    else:
        print('Number is Odd')
        
def welcome_greet():        
    print('Welcome to Nettech!')
    #USern Name?
    name=input('Enter name here ')
    #User Greet
    print('Hello',name)
    print('data Science or Nothing')
    #Course intrest?
    interest=input('Which course are you interested in?')
    #Fees according to course
    print('You chose:',interest)

    if interest=='Data Science':
        print('Fees is 50 Lakh')
    else:
        Print('This course in not yet available!')
        
        
def table():
    user_input=int(input('Enter the number here'))
    for i in range(1,11):
        print(user_input,'x',i,'=',user_input*i)

#Library
import tkinter

#Main Window
window=tkinter.Tk()

#Window size
window.geometry('470x400')

#window title
window.title("Nitesh's Python Portfolio")

#lable UI
tkinter.Label(window,text='PYTHON PROJECTS',font=('Franklin Gothic Medium',18)).place(x=150,y=10)
tkinter.Label(window,text='Made By : Nitesh Chavhan',font=('Times new roman',12)).place(x=165,y=50)
#button -- Start
tkinter.Button(window,text='Factorial',command=factorial_number).place(x=30,y=100,width=130,)
tkinter.Button(window,text='Countdown',command=countdown).place(x=170,y=100,width=130)
tkinter.Button(window,text='Calculate Cube',command=calculate_cube).place(x=310,y=100,width=130)
tkinter.Button(window,text='Dice Game',command=dice_game).place(x=30,y=200,width=130)
tkinter.Button(window,text='Welcome To EPL',command=welcome_to_EPL).place(x=170,y=200,width=130)
tkinter.Button(window,text='Ticket Booking',command=ticket_booking).place(x=310,y=200,width=130)
tkinter.Button(window,text='Odd Even',command=odd_even_number).place(x=30,y=300,width=130)
tkinter.Button(window,text='Welcome Greet',command=welcome_greet).place(x=170,y=300,width=130)
tkinter.Button(window,text='Table',command=table).place(x=310,y=300,width=130)


#ye run nahi hoga to UI nahi dikhega
window.mainloop()
```
